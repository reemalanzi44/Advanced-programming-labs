```python
#Question1
numbers = []
for i in range(5):
    num = float(input(f"Enter number {i+1}: "))
    numbers.append(num)

total = sum(numbers)
average = total / len(numbers)
maximum = max(numbers)
minimum = min(numbers)

print(f"Sum: {total}")
print(f"Average: {average}")
print(f"Maximum: {maximum}")
print(f"Minimum: {minimum}")
```

```python
#Question2
words = []
for i in range(5):
    word = input(f"Enter word {i+1}: ")
    words.append(word)

words = tuple(words)

print(f"First word: {words[0]}")
print(f"Last word: {words[-1]}")
print(f"Total number of words: {len(words)}")

search = input("Enter a word to search: ")
if search in words:
    print(f"{search} is in the tuple")
else:
    print(f"{search} is not in the tuple")
```

```python
## Question3
students = {}

for i in range(3):
    name = input(f"Enter student {i+1} name: ")
    score = float(input(f"Enter {name}'s score: "))
    students[name] = score

print("\nAll students and scores:")
for name, score in students.items():
    print(f"{name}: {score}")

highest = max(students, key=students.get)
print(f"\nStudent with highest score: {highest} with {students[highest]}")
```

```python
#Question4
list1 = input("Enter first list of numbers separated by spaces: ").split()
list2 = input("Enter second list of numbers separated by spaces: ").split()

set1 = set(list1)
set2 = set(list2)

print(f"Union: {set1 | set2}")
print(f"Intersection: {set1 & set2}")
print(f"Difference: {set1 - set2}")
```

```python
#Question5
names = []
for i in range(5):
    name = input(f"Enter name {i+1}: ")
    names.append(name)

names_tuple = tuple(names)
print(f"Tuple: {names_tuple}")

names_dict = {}
for name in names:
    names_dict[name] = len(name)
print(f"Dictionary: {names_dict}")

unique_lengths = set(names_dict.values())
print(f"Unique name lengths: {unique_lengths}")
```

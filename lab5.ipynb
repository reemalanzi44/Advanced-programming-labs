```python
#Question 1
import math

def calculate_area(radius):
    return math.pi * radius ** 2

calculate_circumference = lambda r: 2 * math.pi * r

radius = float(input("Enter the radius of a circle: "))

area = calculate_area(radius)
circumference = calculate_circumference(radius)

print(f"Area: {round(area, 2)}")
print(f"Circumference: {round(circumference, 2)}")
```

```python
#Question 2
import re

def sum_numbers(numbers):
    return sum(numbers)

double = lambda x: x * 2

text = input("Enter a string: ")

found = re.findall(r'\d+\.?\d*', text)
numbers = [float(n) for n in found]

total = sum_numbers(numbers)
doubled = double(total)

print(f"Numbers found: {numbers}")
print(f"Sum: {total}")
print(f"Doubled value: {doubled}")
```

```python
#Question 3
import math

def get_squares(numbers):
    square = lambda x: x ** 2
    return list(map(square, numbers))

nums_input = input("Enter a list of numbers separated by commas: ")
numbers = [float(n) for n in nums_input.split(",")]

squares = get_squares(numbers)

max_square = math.max(squares) if hasattr(math, 'max') else max(squares)
sqrt_sum = math.sqrt(sum(squares))

print(f"Squares: {squares}")
print(f"Maximum square: {max(squares)}")
print(f"Square root of sum of squares: {round(sqrt_sum, 2)}")
```

```python
#Question 4
import re
import math

def calculate_average(scores):
    avg = lambda s: sum(s) / len(s)
    return avg(scores)

password = input("Enter a password: ")

has_upper = re.search(r'[A-Z]', password)
has_lower = re.search(r'[a-z]', password)
has_digit = re.search(r'\d', password)

if has_upper and has_lower and has_digit:
    strength = "strong"
else:
    strength = "weak"

scores_input = input("Enter 5 scores separated by spaces: ")
scores = [float(s) for s in scores_input.split()]

average = calculate_average(scores)
rounded_avg = math.ceil(average)

print(f"Password is: {strength}")
print(f"Rounded average score: {rounded_avg}")
```

```python
#Question 5
import re
import math

def count_vowels(text):
    total = 0
    for char in text:
        if char.lower() in "aeiou":
            total += 1
    return total

paragraph = input("Enter a paragraph of text: ")

vowel_words = re.findall(r'\b[aeiouAEIOU]\w*', paragraph)

total_vowels = count_vowels(paragraph)

word_lengths = list(map(lambda word: len(word), vowel_words))

if len(word_lengths) > 0:
    average_length = round(math.fsum(word_lengths) / len(word_lengths), 2)
else:
    average_length = 0

print(f"\nWords starting with a vowel: {vowel_words}")
print(f"Total vowels in paragraph: {total_vowels}")
print(f"Average vowel-word length: {average_length}")
```

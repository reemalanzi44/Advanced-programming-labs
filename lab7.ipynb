```python
#question1
try:
    num1 = float(input("Enter the first number:"))
    num2 = float(input("Enter the second number:"))
    result = num1 / num2
except ZeroDivisionError:
    print("Error : can Not divide by zero")
except ValueError:
     print("Error: please enter numbers only")
else:
    print("Result:" , result)
    
```

```python
#question2
class Person:
    def __init__(self,name,age):
        self.name = name
        self.__age = age
    def get_age(self):
        return self.__age
    def set_age(self , new_age): 
        self.__age = new_age
    def display(self):
        print("Name:", self.name)
        print("Age:", self.__age)

p = Person("Reem" , 21)
p.set_age(22)
p.display()
```

```python
#question3
class Employee:

    def __init__(self, name, salary):
        self.name = name
        self.salary = salary


class Manager(Employee):

    def add_bonus(self, bonus):
        self.salary = self.salary + bonus

    def display(self):
        print("Name:", self.name)
        print("Total Salary:", self.salary)


m = Manager("reem", 5000)
m.add_bonus(1000)
m.display()

```

```python
#question4

class Student:

    def __init__(self):
        self.name = input("Enter name: ")

        try:
            self.age = int(input("Enter age: "))

            if self.age < 0:
                print("Error: age cannot be negative")
                self.age = None

        except ValueError:
            print("Error: age must be a number")
            self.age = None

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)
s = Student()
s.display()

```

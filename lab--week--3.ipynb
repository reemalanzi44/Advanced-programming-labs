```python
#Question 4
color = input("Enter a color name : red , yellow , green").lower()
match color:
    case "red" :
        print("Action : Stop")
    case "yellow" :
        print("Action : get ready")
    case "green" :
        print("Action : Go")
    case _: 
         print("color invaild!")  
```

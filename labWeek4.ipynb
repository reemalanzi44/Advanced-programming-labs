```python
#Question 1
Total = 0 
for i in range(10): #using for loop to make the user enter 10 num
    number = int(input("Enter a number:"))
    if number < 0: 
        continue
    if number == 0:
        break
    if number % 2 == 0: 
        Total += number
print("the total =", Total)
```

```python
#Question 2
num = int(input("Enter a number:"))
for i in range(1, 13): #for loop from 1 to 12(used 13 cause range stops before the last )
    result = num * i
    if i % 3 == 0:
        continue
    if result > 50:
        break
    print(num , "x" , i , "=" , result)
    
```

```python
#Question 3
SecretNum = 18
Attempts = 0
while True:
    guess = int(input("Guess the number 'Hint(1-20)':"))
    if guess <1 or guess > 20:
        print("Invaild!")
        continue
    Attempts += 1 # to keep track of vaild attempts
    if guess == SecretNum:
        print("Correct, good job!")
        print("number of attampts:" ,Attempts," attempts")
        break   
    
                
                
```

```python
#question 4
n = int(input("Enter a number:"))
count = 0
for i in range ( 1 , n+1): #from 1 to n
    if i > 50:
        break
    if  i % 8 == 0:
        continue
    if  i % 4 == 0:
         count += 1
print("count" , count)
```
